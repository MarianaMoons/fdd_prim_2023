```python

loans['emp_length'] = loans['emp_length'].astype(str)

def definiranios(cell):
    cotasuperior = '10+'
    cotainferior = '< 1'

    if cotasuperior in cell:
        return '10'
    if cotainferior in cell:
        return '0'
    return cell.split()[0]

loans['emp_length'] = loans['emp_length'].apply(lambda x: definiranios(x))
loans['emp_length'].head(10)
```

```python
# Revisemos que no hay valores fuera de los permitidos
# Ya se hicieron pruebas y nos percatamos que hay un cell que tiene un valor no permitido
# con valor 'ANY', indicando un error, por lo que ese valor se pasa a 'OTHER'

loans['home_ownership'] = loans['home_ownership'].astype(str)
def revisarvalores(cell):
    revision = ['RENT', 'MORTGAGE', 'OWN', 'OTHER']
    if cell not in revision:
        cell = 'OTHER'
    return cell
loans['home_ownership'] = loans['home_ownership'].apply(lambda x: revisarvalores(x))
```

Crea codigo para **guardar** y **cargar** un archivo JSON en el que se guarde la `estrategia` y `valor` que utilizaste para **imputar**. Por ejemplo: Si hay una columna que se llama `columna 3` y utilizaste la estrategia de imputacion de media, y existe otra llamada `columna 4` y  elegiste la palabra 'missing' el JSON debera contener:  
  
 `{'columna 3':{'estrategia':'mean', 'valor':3.4}, 'columna 4':{'estrategia':'identificador', 'valor':'missing'}}`  

 De tal manera que para cada columna que tenga un metodo de imputacion apunte a otro diccionario donde el **key** `estrategia` describa de manera sencilla el metodo, y el **key** `valor` el valor usado. En general:   
 `{'nombre de la columna':{'estrategia':'descripcion de estrategia', 'valor':'valor utilizado'}}`. 
 

De utilizar mas de un metodo puedes anidarlos en una lista  
  `[{...},{...}]`.  

Incluso si la columna utilizada no sufrio imputacion, es necesario que la agregues al JSON.

#### Quitar columnas con puros nan's y columnas que no aportan informacion

```python
# Creamos el método para quitar todas las columnas que tengan puros nan's
loans = pd.read_pickle('prestamos.pickle')
#print(loans.head(10))
def quitarColsConPurosNans():
    columnas = loans.columns #Obtenemos todas las columnas para iterar
    result = []
    for columna in columnas:
        if loans[str(columna)].isna().all():
            result.append(str(columna))

    return result

datos_dict = pd.read_excel(
    'https://resources.lendingclub.com/LCDataDictionary.xlsx')
datos_dict.columns = ['feature', 'description']


coldel = []
columnasnas = quitarColsConPurosNans()
for col in columnasnas:
    # si una de las columnas que tiene puros nan's no está en el diccionario con la descripción de cada col lo eliminamos
    if not datos_dict['feature'].eq(str(col)).any():
        coldel.append(str(col))

loans = loans.drop(columns = coldel)



# Queremos ver que la columna unnamed: 0 es igual al indice
serieprueba = loans['Unnamed: 0'] - loans.index

# si la serieprueba es igual a 0 quiere decir que la columna unnamed: 0 es igual al indice
print(serieprueba.eq(0).all())
#También por esta prueba vemos que la columna unnamed: 0 era igual al índice 
# y con esta prueba vemos que no esta en el diccionario con descripción
if not datos_dict['feature'].eq('Unnamed: 0').any():
    loans= loans.drop(columns = ['Unnamed: 0'])


```

#### Imputar valores para columnas que no son fechas

```python
# Para el caso de emp_title podemos ver que para los Nan's podemos poner 'Unspecified'

loans['emp_title'] = loans['emp_title'].fillna('Unspecified')
estrategias = {'emp_title': {'estrategia': 'rellenar con Unspecified', 'valor': 'Unspecified'}}

# Para el caso de emp_length podemos poner 0 para los Nan's, pues alguien que no esta empleado
# no tiene tiempo empleado, por lo que un 0 tiene sentido

loans['emp_length'] = loans['emp_length'].fillna(0)
estrategias['emp_length'] = {'estrategia': 'rellenar con 0', 'valor': 0}

# Para el caso de desc podemos poner no description para los Nan's, pues alguien que no tiene
# descripcion de su prestamo es valido ponerle no description

loans['desc'] = loans['desc'].fillna('no description')
estrategias['desc'] = {'estrategia': 'rellenar con no description', 'valor': 'no description'}

# Para el caso de hardship_loan_status podemos poner no hardship para los Nan's
# pues alguien que no tiene un hardship loan es valido ponerle no hardship

loans['hardship_loan_status'] = loans['hardship_loan_status'].fillna('no hardship')
estrategias['hardship_loan_status'] = {'estrategia': 'rellenar con no hardship', 'valor': 'no hardship'}

# Para el caso de hardship_status podemos poner unstarted para los Nan's
# pues para alguien que nunca ha tenido un hardship es valido ponerle unstarted

loans['hardship_status'] = loans['hardship_status'].fillna('unstarted')
estrategias['hardship_status'] = {'estrategia': 'rellenar con unstarted', 'valor': 'unstarted'}

# Para el caso de hardship_reason podemos poner no reason para los Nan's
# pues para alguien que no tiene hardship reason, poner no reason indicaria que no tiene

loans['hardship_reason'] = loans['hardship_reason'].fillna('no reason')
estrategias['hardship_reason'] = {'estrategia': 'no reason indica no hardship', 'valor': 'no reason'}

# Para el caso de hardship_type podemos poner no hardship para los Nan's

loans['hardship_type'] = loans['hardship_type'].fillna('no hardship')
estrategias['hardship_type'] = {'estrategia': 'no hardship tiene sentido en una columna de puros types', 'valor': 'no hardship'}

# Para el caso de settlement_status podemos poner unstarted para los Nan's
# pues eso es lo que nos indicaria un nan en este caso

loans['settlement_status'] = loans['settlement_status'].fillna('unstarted')
estrategias['settlement_status'] = {'estrategia': 'unstarted puede significar no haber tenido settlement', 'valor': 'unstarted'}


```

#### Eliminar valores para columnas que tengan más de 95% de valores nulos

```python
# Primero vamos a hacer una función que me de el porcentaje de nans que tiene una columna
# si el porcentaje es mayor o igual a 95% la eliminamos

def porcentajeNans(col):
    return loans[str(col)].isna().sum()/len(loans[str(col)])*100

# Ahora vamos a hacer una función que me de las columnas que tienen un porcentaje de nans mayor o igual a 80%

def colsConMuchosNans():
    # Las columnas para iterar son las que tienen fecha y sobre las que hice limpieza
    columnas = {'debt_settlement_flag_date', 'next_pymnt_d', 'last_pymnt_d', 
        'settlement_date', 'hardship_start_date', 'hardship_end_date', 
        'last_credit_pull_d', 'payment_plan_start_date'}
    result = []
    for columna in columnas:
        if porcentajeNans(str(columna)) >= 95:
            result.append(str(columna))

    return result
colsparaeliminar = colsConMuchosNans()

# Eliminamos estas columnas 

loans = loans.drop(columns = colsparaeliminar)

# Escribimos la estrategia para eliminar estas columnas

for col in colsparaeliminar:
    estrategias[col] = {'estrategia': 'eliminar porque el porcentaje de Nan era mayor a 95%', 'valor': ''}
```
